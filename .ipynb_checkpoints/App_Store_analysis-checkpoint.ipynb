```python
df_apps.Installs = df.Installs.astype(str).str.replace(',', '')
df_apps.Installs = pd.to_numeric(df.Installs)
df_apps[['App', 'Installs']].groupby('Installs').count()
```

```python
df.Price
```

```python
df_apps.Price = df.Price.astype(str).str.replace('$', '')
df_apps.Price = pd.to_numeric(df_apps.Price)
df_apps[['App', 'Price']].sort_values(by='Price', ascending=False).head(20)
```

```python
df_apps = df_apps[df_apps.Price < 250]
df_apps.Price = pd.to_numeric(df_apps.Price)
df_apps.sort_values(by='Price', ascending=False).head(20)
```

```python
df_apps['Revenue_Estimate'] = df_apps.Price.mul(df_apps.Installs)
df_apps.sort_values(by='Revenue_Estimate', ascending=False).head(10)
```

```python
df_apps.Category.nunique()
```

```python
top10_category = df_apps.Category.value_counts()[:10]
top10_category
```

```python
bar = px.bar(x=top10_category.index, y=top10_category.values, title='Number of Category')
bar.update_layout(yaxis_title='Count', xaxis_title='Category')
bar.show()
```

```python
category_installs = df_apps.groupby('Category').agg({ 'Installs' : pd.Series.sum }).sort_values('Installs', ascending=False)
category_installs
```

```python
h_bar = px.bar(x=category_installs.Installs, y=category_installs.index, orientation='h', title='Category Popularity')
h_bar.update_layout(yaxis_title='Number of Installs', xaxis_title='Category')
h_bar.show()
```

```python
category_number = df_apps.groupby('Category').agg({ 'App': pd.Series.count }).sort_values('App', ascending=False)
category_number
```

```python
category_merged = pd.merge(category_installs, category_number, on='Category', how='inner')
print(f'The dimentions of the DataFrame are: {category_merged.shape}')
category_merged.sort_values('Installs', ascending=False)
```

```python
scatter = px.scatter(category_merged, 
                    x='App',
                    y='Installs',
                    title='Category Concentration',
                    hover_name=category_merged.index,
                    color='Installs')
scatter.update_layout(xaxis_title='Number of Apps (Lower=More concentrated)',
                     yaxis_title='Installs',
                     yaxis=dict(type='log'))
scatter.show()
```

```python
df_apps.Genres.nunique()
```

```python
df_apps.Genres.value_counts()
```

```python
stack = df_apps.Genres.str.split(';', expand=True).stack()
print(f'We now have a single column with shape: {stack.shape}')
num_genres = stack.value_counts()
print(f'Number of genres: {len(num_genres)}')
```

```python
num_genres
```

```python
bar = px.bar(x=num_genres.index,
             y=num_genres.values,
            title='Number of genres',
            hover_name=num_genres.index,
            color=num_genres.values,
            color_continuous_scale='Agsunset')
bar.update_layout(yaxis_title='Number of Apps', xaxis_title='Genres', coloraxis_showscale=False)

bar.show()
```

```python
df_free_vs_paid = df_apps.groupby(['Category', 'Type'], as_index=False).agg({ 'App': pd.Series.count })
df_free_vs_paid
```

```python
g_bar = px.bar(df_free_vs_paid,
               x=df_free_vs_paid.Category,
              y=df_free_vs_paid.App,
              color='Type',
              barmode='group',
              title='Free vs Paid Apps by Category')
g_bar.update_layout(xaxis_title='Category',
                   yaxis_title='Number of Apps',
                   xaxis={'categoryorder': 'total descending'},
                   yaxis=dict(type='log')
                   )
g_bar.show()
```

```python
box = px.box(df_apps,
            x='Type',
            y='Installs',
            color='Type',
            points='all',
            notched=True,
            title='How Many Downloads are Paid Apps Giving Up?')
box.update_layout(yaxis=dict(type='log'))
box.show()
```

```python
df_paid_apps = df_apps[df_apps.Type == 'Paid']
df_paid_apps
```

```python
box = px.box(df_paid_apps,
            x='Category',
            y='Revenue_Estimate',
            title='How much Can Paid Apps Earn?')

box.update_layout(xaxis_title='Category',
                  yaxis_title='Paid Apps BallPark Revenue',
                  xaxis={'categoryorder': 'min ascending'},
                  yaxis=dict(type='log'))
box.show()
```

```python
box = px.box(df_paid_apps,
            x='Category',
            y='Price',
            title='Price Per Category')

box.update_layout(xaxis_title='Category',
                  yaxis_title='Paid Apps Price',
                  xaxis={'categoryorder': 'max descending'},
                  yaxis=dict(type='log'))
box.show()
```
