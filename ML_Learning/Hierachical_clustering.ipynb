```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./Datasets/Mall_Customers.csv')
X = data.iloc[:, [3, 4]].values
data
```

## Using denograms to get the optimal number of clusters

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))  # method = 'ward' reduces or minimizes the variance between clusters
plt.title('Dendrogram')
plt.xlabel('Customers (Observation points)')
plt.ylabel('Euclidean distances')
plt.show()
```

## Training the Hierachical clustering Model on the dataset

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

```python
y_hc
```

## Visualizing the clusters

```python
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 100, label = 'Cluster 1', c = 'red')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 100, c ='blue', label = 'Cluster 2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 100, c ='green', label = 'Cluster 3')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], s = 100, c ='orange', label = 'Cluster 4')
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], s = 100, c ='cyan', label = 'Cluster 5')
plt.title('Clusters of customers')
plt.xlabel('Anual Income (k$)')
plt.ylabel('Spending score (1 - 100)')
plt.legend()
plt.show()
```
