```python
# Convert the datatype to datetime
data['DATE'] = pd.to_datetime(data['DATE'])
data.head()
```

## Get the number of posts of each language in a particular month

```python
reshaped_df = data.pivot(columns='TAGS', index='DATE', values='POSTS')
reshaped_df
```

```python
reshaped_df.count()
```

```python
reshaped_df.fillna(0, inplace=True)
```

```python
reshaped_df
```

```python
reshaped_df.isna().values.any()
```

```python
plt.figure(figsize=(16, 8))
plt.xlabel('DATE', fontsize=16)
plt.ylabel('POSTS', fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.ylim(0, 35000)

for column in reshaped_df.columns:
    plt.plot(reshaped_df.index, reshaped_df[column], label=reshaped_df[column].name, linewidth=3)
    
plt.legend(fontsize=16)
```

```python
# To remove noisiness from the data, I'll construct the rolling mean by reducing the number of observations and 
# computing the mean
# The window is number of observations that are averaged

rolled_df = reshaped_df.rolling(window=5).mean()

plt.figure(figsize=(16, 8))
plt.xlabel('DATE', fontsize=16)
plt.ylabel('Number of posts', fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.ylim(0, 35000)

for column in rolled_df.columns:
    plt.plot(rolled_df.index, rolled_df[column], linewidth=3, label=rolled_df[column].name)
    
plt.legend(fontsize=16)
```
