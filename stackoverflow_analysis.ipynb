```python
# Convert the datatype to datetime
data['DATE'] = pd.to_datetime(data['DATE'])
data.head()
```

## Get the number of posts of each language in a particular month

```python
reshaped_df = data.pivot(columns='TAGS', index='DATE', values='POSTS')
reshaped_df
```

```python
reshaped_df.count()
```

```python
reshaped_df.fillna(0, inplace=True)
```

```python
reshaped_df
```

```python
reshaped_df.isna().values.any()
```

```python
plt.figure(figsize=(16, 8))
plt.xlabel('DATE', fontsize=16)
plt.ylabel('POSTS', fontsize=16)
plt.ylim(0, 35000)
plt.plot(reshaped_df.index, reshaped_df.java)
```
