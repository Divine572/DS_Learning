```python
img = misc.face()
plt.imshow(img)
```

```python
img.shape
```

```python
type(img)
```

```python
# Converting an image to grayscale
# Y(linear) = 0.2126R(linear) + 0.7152G(linear) + 0.0722B(linear)
sRGB_array = img / 255
grey_vals = np.array([0.2126, 0.7152, 0.0722])
image_grey = sRGB_array @ grey_vals
image_grey
```

```python
plt.imshow(image_grey, cmap = 'gray')
```

```python
plt.imshow(np.flip(image_grey), cmap = 'gray')
```
